```python
from pprint import pprint as print
from gensim.models.fasttext import FastText as FT_gensim
from gensim.test.utils import datapath
from gensim.models.word2vec import Word2Vec, Text8Corpus
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
path = '/Users/stefanoraimondousai/Documents/ReadingCourse/WikiPulito.txt'
sentences = datapath(path)
```

```python
model = Word2Vec(size=300, min_count=250, sg=1, workers=8, negative=10)
```

```python
model.build_vocab(corpus_file=sentences, progress_per=50000)
```

```python
model.train(
    corpus_file=sentences, epochs=5,
    total_examples=model.corpus_count, total_words=model.corpus_total_words
)
```

```python
model.save('WikiW2C300neg10ep5.model')
```
