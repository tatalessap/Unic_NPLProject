```python
%%time
documents = {'Marta Pibiri' : [], 'Stefano Raimondo Usa' : [] }
marta = 'Marta Pibiri'
stefano = 'Stefano Raimondo Usa'
flag = 0
lista=[]
```

```python
%%time
for f in file: 
    soup = BeautifulSoup(open(f'./ChatExport_05_03_2020 2/{f}'), "html.parser")
    for i,div in enumerate(soup.findAll("div")):

        if marta in div.text:
            flag = 1
        if stefano in div.text:
            flag = 0

        t = re.sub(r'(?<!\.)\n', '', div.text)
        t = re.sub(r'       ', '', t)
        #lista.append(div)
        if div.attrs.get('class')[0] == 'text': 
            documents.get(marta).append(t) if flag == 1 else documents.get(stefano).append(t)

    
```
